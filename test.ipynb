```python
def top_k(items, k):
    return sorted(items, key=lambda x: x[1], reverse=True)[:k]
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] += 1
    return d
```

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) // 2
    return total - sum(nums)
```

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)
```

```python
def moving_average(stream, w):
    result = []
    window_sum = 0

    for i, val in enumerate(stream):
        window_sum += val

        if i >= w:
            window_sum -= stream[i - w]

        result.append(window_sum / min(i + 1, w))

    return result
```
