```python
class Student:
    def __init__(self,name,sap_id,phy,chem,maths):
        self.name=name
        self.sap_id=sap_id
        self.marks=[phy,chem,maths]

    def Display(self):
        print(self.name,self.sap_id,self.marks)

    def Marks_percentage(self):
        return sum(self.marks)/3

    def Result(self):
        if all(m>=40 for m in self.marks):
            return "Pass"
        else:
            return "Fail"

def class_average(students):
    return sum(s.Marks_percentage() for s in students)/len(students)

n=int(input("Enter number of students: "))
students=[]
for i in range(n):
    name=input("Enter name: ")
    sap_id=input("Enter SAP ID: ")
    phy=int(input("Enter Physics marks: "))
    chem=int(input("Enter Chemistry marks: "))
    maths=int(input("Enter Maths marks: "))
    students.append(Student(name,sap_id,phy,chem,maths))

for s in students:
    s.Display()
    print("Percentage:",s.Marks_percentage())
    print("Result:",s.Result())

print("Class Average:",class_average(students))
```

```python

class A:
    def show(self):print("Class A")
class B(A):
    def display(self):print("Class B")
b=B();b.show();b.display()


class C:
    def cshow(self):print("Class C")
class D:
    def dshow(self):print("Class D")
class E(C,D):
    def eshow(self):print("Class E")
e=E();e.cshow();e.dshow();e.eshow()


class F:
    def fshow(self):print("Class F")
class G(F):
    def gshow(self):print("Class G")
class H(G):
    def hshow(self):print("Class H")
h=H();h.fshow();h.gshow();h.hshow()
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,other):
        return Point(self.x+other.x,self.y+other.y)
    def __str__(self):
        return f"({self.x},{self.y})"

P1=Point(10,20)
P2=Point(12,15)
P3=P1+P2
print(P3)
```
