```python
class student:
    school_name="UPES"
    def __init__(self,name,age):
        self.name=name
        self.age=age

s1=student("Aman",20)
s2=student("Riya",21)

print("School:", s1.school_name)
print("School:", s2.school_name)

print("Student 1:", s1.name, s1.age)
print("Student 2:", s2.name, s2.age)
```

```python
class employee:
    company_name="ABC Ltd"
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def show_details(self):
        print(self.name, self.salary, employee.company_name)

    @classmethod
    def change_company_name(cls, new_name):
        cls.company_name=new_name

e1=employee("Aman", 30000)
e2=employee("Riya", 40000)

print("Before change")
e1.show_details()
e2.show_details()

employee.change_company_name("XYZ Ltd")

print("After change")
e1.show_details()
e2.show_details()
```

```python
class mathhelper:

    @staticmethod
    def is_even(num):
        return num % 2 == 0
    
print(mathhelper.is_even(8))
print(mathhelper.is_even(11))
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"product : {self.name}, price : {self.price}"
    
    def __add__(self, other):
        return self.price + other.price
    
    def __eq__(self, other):
        return self.price == other.price


# Create objects OUTSIDE the class
p1 = Product("Laptop", 50000)
p2 = Product("Phone", 30000)
p3 = Product("Tablet", 50000)

print(p1)
print(p1 + p2)
print(p1 == p3)
print(p1 == p2)
```
