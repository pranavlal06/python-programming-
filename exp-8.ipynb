```python
with open("name.txt","w") as f:
    f.write("Alice\nEve\nOscar\nUma\nIan\nBob\nCharlie\n")

with open("name.txt") as f:
    names=f.read().splitlines()
print(len(names))
print(sum(1 for n in names if n[0].lower() in "aeiou"))
print(max(names,key=len))
```

```python
with open("numbers.txt","w") as f:
    f.write("45\n120\n300\n78\n99\n")

with open("numbers.txt") as f:
    nums=list(map(int,f.read().splitlines()))
print(max(nums))
print(sum(nums)/len(nums))
print(sum(1 for n in nums if n>100))
```

```python
with open("city.txt","w") as f:
    f.write("Dehradun 5.78 308.20\nDelhi 190 1484\nMumbai 125 603\nChennai 70 426\nLucknow 28 349\n")

with open("city.txt") as f:
    cities=[line.split() for line in f]
for c in cities:
    print(c[0],c[1],c[2])
print([c[0] for c in cities if float(c[1])>10])
print(sum(float(c[2]) for c in cities))
```

```python
data=["1 0","2 $","3 1"]
for line in data:
    try:
        a,b=line.split()
        print(int(a)//int(b))
    except ZeroDivisionError as e:
        print("Error Code:",e)
    except ValueError as e:
        print("Error Code:",e)
```

```python
class FileEmptyError(Exception):pass
class FileFormatError(Exception):pass

def check_file(fname):
    with open(fname) as f:
        content=f.read().strip()
        if not content:
            raise FileEmptyError("File is empty")
        if any(ch.isalpha() for ch in content):
            raise FileFormatError("File contains invalid characters")
    return "File OK"

try:
    print(check_file("numbers.txt"))
except (FileEmptyError,FileFormatError) as e:
    print("Custom Exception:",e)
```
